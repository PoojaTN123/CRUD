# **Matplotlib**

# Matplotlib is for data visulization

```python
import matplotlib.pyplot as plt
```

```python
# hours vs marks
Hours = [1,2,3,4,5]
mathMarks = [10,20,30,40,50]
scienceMarks = [20,40,60,80,100]
plt.scatter(Hours,mathMarks)
plt.scatter(Hours,scienceMarks)
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.title("Hours vs Marks")
plt.legend(["Math","Science"])
plt.show()
```
